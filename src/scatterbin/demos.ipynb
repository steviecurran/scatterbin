```python
import numpy as np
import pandas as pd
import scatterbin
```

### Small data set of 123 entries

```python
data = pd.read_csv("hw_all-123.dat",delim_whitespace=True, header=None)
```

**Default settings**

```python
scatterbin.plot(data)
```

**Quantile binning and showing p-value**

```python
scatterbin.plot(data, pvalue=True, equal_span = 'n')
```

**Quantile binning showing the x-range and including stray points**

```python
scatterbin.plot(data, equal_span = 'nx',inc_strays = True)
```

## Larget dataset with 5000 entries

```python
data = pd.read_csv("hw_all.dat",delim_whitespace=True, header=None)
```

```python
scatterbin.plot(data,fs = 14,xlabel = "Height [cm]",ylabel = "Weight [kg]",
              pvalue=True, equal_span = 'nx',point_fc = 'b',point_ec = 'b',
              point_s = 2, nbins = 10, two_panel = True, width=7,height=7)
```
